```python
import pandas as pd

# Load the dataset
file_path = 'NYC_Planimetric_Database__Sidewalk_20250408.csv'
df = pd.read_csv(file_path)

# Columns to drop
columns_to_drop = ['SOURCE_ID', 'SUB_CODE', 'FEAT_CODE']

# Drop columns (working directly with df)
df_cleaned = df.drop(columns=[col for col in columns_to_drop if col in df.columns])

# Display rows
print(df_cleaned.sample(5))
df_cleaned.head(5)
```

```python
import pandas as pd
import geopandas as gpd
from shapely import wkt

# Load the cleaned dataset
df = pd.read_csv('Cleaned_NYC_Planimetric_Database__Sidewalk_20250408.csv')  # use your actual cleaned file name

# Convert 'the_geom' from WKT to actual geometry
df['geometry'] = df['the_geom'].apply(wkt.loads)
gdf = gpd.GeoDataFrame(df, geometry='geometry')
gdf.set_crs(epsg=4326, inplace=True)

# Drop the original WKT column (optional but cleaner)
gdf = gdf.drop(columns=['the_geom'])

# Export to GeoJSON
gdf.to_file('Cleaned_NYC_Planimetric_Database__Sidewalk_20250408.geojson', driver='GeoJSON')
```
