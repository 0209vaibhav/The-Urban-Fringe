```python
import pandas as pd

# Load the dataset
file_path = 'Open_Restaurants_Inspections_20250408.csv'  # Update path if needed
df = pd.read_csv(file_path)

# Filter to only include restaurants in Manhattan
df_manhattan = df[df['Borough'] == 'Manhattan']

# Columns to drop
columns_to_drop = [
    'IsSidewayCompliant',
    'AgencyCode',
    'SkippedReason',
    'InspectedOn',
    'IsRoadwayCompliant',
    'CommunityBoard',
    'CouncilDistrict',
    'CensusTract',
    'BIN',
    'BBL'
]


# Drop the specified columns if they exist
df_manhattan_cleaned = df_manhattan.drop(columns=[col for col in columns_to_drop if col in df_manhattan.columns])

# Display 10 random rows
print(df_manhattan_cleaned.sample(5))
df_manhattan_cleaned.head(5)
```

```python
# Save the cleaned DataFrame to a new CSV file
df_manhattan_cleaned.to_csv('Cleaned_Open_Restaurants_Inspections_20250408.csv', index=False)
print("Exported successfully!")
```

```python
import pandas as pd
import json

# Load the inspections dataset
df = pd.read_csv("Cleaned_Open_Restaurants_Inspections_20250408.csv")

# Drop rows without coordinates
df_clean = df.dropna(subset=["Latitude", "Longitude"])

# Convert each row into a GeoJSON Feature
def row_to_feature(row):
    return {
        "type": "Feature",
        "geometry": {
            "type": "Point",
            "coordinates": [row["Longitude"], row["Latitude"]],
        },
        "properties": {
            "restaurant_name": row["RestaurantName"],
            "borough": row["Borough"],
            "address": row["BusinessAddress"],
            "legal_name": row["LegalBusinessName"],
            "seating_choice": row["SeatingChoice"],
            "inspection_id": row["RestaurantInspectionID"],
            "nta": row["NTA"]
        }
    }

# Convert the DataFrame to a list of GeoJSON features
features = [row_to_feature(row) for _, row in df_clean.iterrows()]

# Create the final GeoJSON structure
geojson = {
    "type": "FeatureCollection",
    "features": features
}

# Save to file
with open("Open_Restaurants_Inspections_20250408.geojson", "w") as f:
    json.dump(geojson, f, indent=2)

print("GeoJSON file created as 'Open_Restaurants_Inspections_20250408.geojson'")
```
