```python
import pandas as pd

# Load the dataset
file_path = 'Open_Restaurant_Applications__Historic__20250408.csv'  # Update path if needed
df = pd.read_csv(file_path)

# Filter to only include restaurants in Manhattan
df_manhattan = df[df['Borough'] == 'Manhattan']

# Columns to drop
columns_to_drop = [
    'Food Service Establishment Permit #',
    'Qualify Alcohol',
    'Time of Submission',
    'SLA Serial Number', 'SLA License Type',
    'Landmark District or Building',
    'landmarkDistrict_terms',
    'healthCompliance_terms',
    'Community Board',
    'Council District',
    'Census Tract',
    'BIN',
    'BBL'
]


# Drop the specified columns if they exist
df_manhattan_cleaned = df_manhattan.drop(columns=[col for col in columns_to_drop if col in df_manhattan.columns])

# Display 10 random rows
print(df_manhattan_cleaned.sample(5))
df_manhattan_cleaned.head(5)
```

```python
# Save the cleaned DataFrame to a new CSV file
df_manhattan_cleaned.to_csv('Cleaned_Open_Restaurant_Applications__Historic__20250408.csv', index=False)
print("Exported successfully!")
```

```python
import pandas as pd
import json

# Load the dataset
df = pd.read_csv("Cleaned_Open_Restaurant_Applications__Historic__20250408.csv")

# Function to convert a row into a GeoJSON Feature
def row_to_feature(row):
    return {
        "type": "Feature",
        "geometry": {
            "type": "Point",
            "coordinates": [row["Longitude"], row["Latitude"]],
        },
        "properties": {
            "restaurant_name": row["Restaurant Name"],
            "borough": row["Borough"],
            "street": row["Street"],
            "seating_type": row["Seating Interest (Sidewalk/Roadway/Both)"],
            "approved_sidewalk": row["Approved for Sidewalk Seating"],
            "approved_roadway": row["Approved for Roadway Seating"],
            "nta": row["NTA"]
        }
    }

# Remove rows with missing coordinates
df_clean = df.dropna(subset=["Latitude", "Longitude"])

# Convert each row into a GeoJSON Feature
features = [row_to_feature(row) for _, row in df_clean.iterrows()]

# Create the FeatureCollection
geojson = {
    "type": "FeatureCollection",
    "features": features
}

# Save to file
with open("restaurants.geojson", "w") as f:
    json.dump(geojson, f, indent=2)

print("GeoJSON file created as 'restaurants.geojson'")
```
