```python
import pandas as pd
import geopandas as gpd
from shapely import wkt

# Load CSV
df = pd.read_csv("SIDEWALK_LINE_20250408.csv")

# Convert WKT geometries to Shapely objects
df["geometry"] = df["the_geom"].apply(wkt.loads)

# Drop the original 'the_geom' column
df = df.drop(columns=["the_geom"])

# Create GeoDataFrame
gdf = gpd.GeoDataFrame(df, geometry="geometry", crs="EPSG:4326")

# Export to GeoJSON
gdf.to_file("sidewalk_lines.geojson", driver="GeoJSON")
```
